## Setup

```python
# imports and dependenceis
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# load data into df
crypto_df = pd.read_csv('Resources/crypto_market_data.csv', index_col='coin_id')
crypto_df.head(10)
```

## Scale Data
---

```python
# save column names
col_names = list(crypto_df.columns)

# normalize data
scaled_cols = StandardScaler().fit_transform(crypto_df[col_names])

# create new df
scaled_df = pd.DataFrame(
    scaled_cols,
    columns=col_names,
    index=crypto_df.index
)

# display
scaled_df.head()
```
