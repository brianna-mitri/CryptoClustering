## Functions: K Means
---

### Get elbow plot

```python
# function for elbow plot
def get_elbow_plot(main_df):
    # get copy of df
    df = main_df.copy()

    # get inertias for k vals
    inertias = []
    k_vals = list(range(1, 11))

    for k in k_vals:
        # fit k means model
        k_model = KMeans(n_clusters=k, random_state=0)
        k_model.fit(df)

        # save inertia
        inertias.append(k_model.inertia_)

    # save to df
    elbow_df = pd.DataFrame({
        'k': k_vals, 
        'inertia': inertias
    })

    # plot
    elbow_plot = elbow_df.hvplot.line(
        x='k',
        y='inertia',
        xticks=k
    )
    
    # return 
    return elbow_plot
```

### Perform k means with best k value

```python
# define function to get best k value with CH score and perform k means
def get_k_means_predictions(main_df, k_list, x_val='x', y_val='y'):
    # get copy of df
    df = main_df.copy()

    # -----------------
    # find best k value
    # -----------------
    # score the two k values
    k_scores = []

    for k in k_list:

        # get scores
        k_model = KMeans(n_clusters=k, random_state=0)
        k_model.fit(df)
        labels = k_model.labels_
        score = metrics.calinski_harabasz_score(df, labels)
        k_scores.append({
            'k': k,
            'score': score
        })
        
    # display scores
    print(k_scores)

    # get best k value
    k_val = max(k_scores, key=lambda x: x['score'])['k']
    print(f'Best k value: {k_val}')

    # -------------------
    # perform k means
    # -------------------
    # build k means model with best k value
    k_model = KMeans(n_clusters=k_val, random_state=0)
    k_model.fit(df)

    # predict model clusters
    predictions = k_model.predict(df)

    # add predictions to df
    predictions_df = df.copy()
    predictions_df['predictions'] = predictions

    # -------------------
    # get scatter plot
    # -------------------
    scatter_plot = predictions_df.hvplot.scatter(
        x=x_val,
        y=y_val,
        by='predictions',
        hover_cols='coin_id'
    )

    # return df
    return predictions_df, scatter_plot
```

## K Means with Full Scaled Data
---

### Elbow method (between k=4 and k=5)

```python
# perform elbow method --> between k=4 and k=5
scaled_elbow_plot = get_elbow_plot(scaled_df)
scaled_elbow_plot
```

### Get best k val with CH score and cluster--> k=4

```python
# set up for function
k_options = [4, 5]  #between k=4 and k=5
x_val = 'price_change_percentage_24h'
y_val = 'price_change_percentage_7d'

# peform function
scaled_predictions_df, scaled_predictions_plot = get_k_means_predictions(scaled_df, k_options, x_val=x_val, y_val=y_val)

# display df and plot
scaled_predictions_df.head()
```

```python
# display plot
scaled_predictions_plot
```
