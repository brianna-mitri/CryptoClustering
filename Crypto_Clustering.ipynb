```python
# imports and dependenceis
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
# load data into df
crypto_df = pd.read_csv('Resources/crypto_market_data.csv', index_col='coin_id')
crypto_df.head(10)
```

## Scale Data
---

```python
# save column names
col_names = list(crypto_df.columns)

# normalize data
scaled_cols = StandardScaler().fit_transform(crypto_df[col_names])

# create new df
scaled_df = pd.DataFrame(
    scaled_cols,
    columns=col_names,
    index=crypto_df.index
)

# display
scaled_df.head()
```

## K Means with Full Scaled Data
---

### Find best k value

##### Elbow method

```python
# lists for looping
inertias = []
k_vals = list(range(1, 11))

# get intertias for each k value
for k in k_vals:
    # fit k means model
    k_model = KMeans(n_clusters=k, random_state=0)
    k_model.fit(scaled_df)

    # save inertia
    inertias.append(k_model.inertia_)
```

```python
# save to df
elbow_df = pd.DataFrame({
    'k': k_vals, 
    'inertia': inertias
})

# plot
elbow_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k
)
```

##### Between k=4 and k=5--> test with CH Score

```python
# score the two k values
for k in [4, 5]:
    k_model = KMeans(n_clusters=k, random_state=0)
    k_model.fit(scaled_df)
    labels = k_model.labels_
    score = metrics.calinski_harabasz_score(scaled_df, labels)
    print(f'k ({k}): {score}')
```

##### Best value for k--> 4

### Clustering

```python
# build k means model with best k value
k_scaled_model = KMeans(n_clusters=4, random_state=0)
k_scaled_model.fit(scaled_df)

# predict model clusters
scaled_clusters = k_scaled_model.predict(scaled_df)

# add predictions to df
predictions_scaled_df = scaled_df.copy()
predictions_scaled_df['Cluster_k4'] = scaled_clusters

# review df
predictions_scaled_df.head()
```

```python
# create a scatter plot 
predictions_scaled_df.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Cluster_k4',
    hover_cols='coin_id'
)
```
